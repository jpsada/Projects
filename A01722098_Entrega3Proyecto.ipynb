```python
preguntas = [
           ['1', 'Matemáticas', '3*3', '6', '9', '1', '12', '9' ],
           ['2', 'Ciencias', '¿Cuál es un animal?', 'Lobo', 'Lapiz', 'Pino', 'Arresife', 'Lobo' ],
           ['3', 'Ciencias', 'Texto de la pregunta', 'R1', 'R2', 'R3', 'R4', 'RC' ]]

#preguntas = []
import time

## En esta función se le pide al usuario que ingrese los valores necesarios para poder llenar la pregunta.
# Se pide el número de pregunta, el área de estudio, la pregunta, posibles respuestas y la respuesta correcta.
# Admemás la información ingresada se guarda en una lista y se añade a la lista de preguntas que se muestra arriba.
# Ya que el código se ejecuta exitosamente, se le agradece al usuario.
def registrar_pregunta():

  global preguntas

  print("Bienvenido a la opción para registar tus preguantas, sigue las siguientes instrucciones para guardar las preguntas")

  número_de_pregunta = int(input("Ingresa el número (#) de pregunta: "))
  área_de_estudio = input("Ingresa el área de tu pregunta (Lectura, Matemáticas, Ciencias): ")
  pregunta = input("Ingresa la pregunta: ")
  respuestas = ""

  for num in range(4):
    respuesta = input("Ingresa la respuesta " + str(num + 1) + ": ")

    if num == 3:
      respuestas = respuestas + respuesta # string de respuesta 4, no lleva coma
    else:
      respuestas = respuestas + respuesta + "," # string de respuestas 1, 2, 3

  respuestas = respuestas.split(",") # hacer una lista desde un string, cada elemento es separado por coma

  respuesta1, respuesta2, respuesta3, respuesta4 = respuestas # asignación múltiple de variables de una lista (elemento por elemento)

  correcta = input("Ingresa la respuesta correcta: ")

  pregunta = [número_de_pregunta, área_de_estudio, pregunta, respuesta1, respuesta2, respuesta3, respuesta4, correcta]

  preguntas.append(pregunta)

  leer_archivo()

  print("¡Tu pregunta ha sido registrada exitosamente!")


## En esta función se muestra la lista en donde se encuentran las lista en donde se guardo la información de la pregunta ingresada por el usuario.
# Se da la opción de decidir si se desea cambiar alguna pregunta y al elegir que sí, se le es posible al usuario cambiar la pregunta seleccionada, si elige que no, no entra al proceso.
# Se usa un ciclo for para poder llevar el conteo de la pregunta que el usario selecciono.
# Se confirma que el ID que el usario inserto si sea existente y poder seguir.
# Se le pide al usuario que ingrese la nueva pregunta, poniendo en un orden respectivo, la área, la pregunta, las cuatro posibles respuestas y la respuesta correcta.
# Se imprime la nueva pregunta, ya que fue cambiada y se pregunta si el usuario desea confirmar el cambio, ya que se hace el cambio, se imprime que la pregunta se guardo, si el usuario no seasea guardarla, no se guarda.
def actualizar_pregunta():
  
  global preguntas

  print("Bienvenido a la opción para actualizar preguntas, por favor sigue las instrucciones \nA continuación se despliega la lista de preguntas guardadas\n")

  for pregunta in preguntas:
    
    número_de_pregunta, área_de_estudio, pregunta, respuesta1, respuesta2, respuesta3, respuesta4, correcta = pregunta # esto (pregunta) es una lista, hacemos asignación múltiple

    print(f"ID de la pregunta: {número_de_pregunta} \nEnfoque de la pregunta: {área_de_estudio} \nPregunta: {pregunta}")
    print(f"a){respuesta1:<15}b){respuesta2:<15}c){respuesta3:<15}d){respuesta4:<15} \n")
  
  confirmación_cambiar_pregunta = input("¿Deseas cambiar alguna pregunta? (Sí/No): ")

  if confirmación_cambiar_pregunta == 'Sí' or confirmación_cambiar_pregunta == 'sí' or confirmación_cambiar_pregunta == 'Si' or confirmación_cambiar_pregunta == 'si':
   
    id_número_cambio =  int(input("Ingresa el id de la pregunta que quieres cambiar: "))

    if id_número_cambio >= 1 and id_número_cambio <= len(preguntas):

      print(preguntas[id_número_cambio - 1])

      cambio_de_pregunta = preguntas[id_número_cambio - 1]


    for i in range(1, len(cambio_de_pregunta)):

      #print(cambio_de_pregunta[i])

      nueva_pregunta = input("Ingresa los nuevos datos de tu pregunta (Área, pregunta, las cuatro opciones y la opción correcta): ")

      if nueva_pregunta != "":

        cambio_de_pregunta[i] = nueva_pregunta


    print(cambio_de_pregunta)

    actualizar_pregunta = input("¿Deseas confirmar la actualicación de tu pregunta? (Sí/No)")


    if actualizar_pregunta == 'Sí' or actualizar_pregunta == 'sí' or actualizar_pregunta == 'Si' or actualizar_pregunta == 'si':
     
      preguntas[id_número_cambio - 1] = cambio_de_pregunta
     
      print(preguntas[id_número_cambio - 1])
     
      print("Los cambios fueron realizados exitosamente.")

    else:

      print("Ningun cambio fue realizado.")

  else:

    print("El ID ingresa no fue encontrado, favor de meter un ID válido.")




nombre_guardar_archivo = "preguntas_guardadas.txt"

def guardar_archivo(nombre):

  global preguntas
  global pregunta

  with open(nombre, "w") as archivo:

    for pregunta in preguntas:
      pregunta[1] = str(pregunta[1])

      texto = ", ".join(pregunta)
      archivo.write(texto + "\n")

guardar_archivo(nombre_guardar_archivo)


## En esta función, se le pide al usuario que seleccione el área de estudio para poder practicar dentro de esa área especifica.
# Ya que se elige el área de estudio, se muestran las pregutnas de esa área en especifico, junto con sus posibles respuestas.
# El usuario puede insertar su respuesta y se le muestra si su respuesta fue correcta o incorrecta.
# Al finalizar con las preguntas del área seleccionada, se muestra la calificación obtenida en el estudio.
def estudiar_área():
  
  lista_área = []
  contador_preguntas_totales_estudio = 0
  contador_preguntas_correctas_estudio = 0

  área_para_estudiar = input("Ingresa el área de estudio que quieres estudiar (Lectura, Matemáticas, Ciencias): ")

  for pregunta in preguntas:

    if pregunta[1] == área_para_estudiar:

      lista_área.append(pregunta)


  for pregunta in lista_área:
    
    número_de_pregunta, área_de_estudio, pregunta, respuesta1, respuesta2, respuesta3, respuesta4, correcta = pregunta # esto (pregunta) es una lista, hacemos asignación múltiple

    print(f"ID de la pregunta: {número_de_pregunta} \nEnfoque de la pregunta: {área_de_estudio} \nPregunta: {pregunta}")
    print(f"a){respuesta1:<15}b){respuesta2:<15}c){respuesta3:<15}d){respuesta4:<15} \n")

    contador_preguntas_totales_estudio = contador_preguntas_totales_estudio + 1

    respuesta_estudio = input("Ingresa la respuesta correcta: ")

    if respuesta_estudio == correcta:

      print("\nRespuesta correcta\n")

      contador_preguntas_correctas_estudio = contador_preguntas_correctas_estudio + 1

    else:
      print("\nRespuesta incorrecta\n")

  calificación_estudio = (contador_preguntas_correctas_estudio / contador_preguntas_totales_estudio) * 100
  
  print(f'La calificación de tu estudio en el área de {área_para_estudiar} es de {calificación_estudio}\n')

## En esta función se muestras las preguntas dentro de la lista original, en donde se le pide al usuario que lea las preguntas y conteste de manera correcta la pregunta.
# Se lleva un conteo de el número de preguntas que son contestadas y del número de preguntas que son contestadas correctamente.
# Se hace el cálculo de la califación del usuario y se guarda directamente en la función de "reportar_calificaciones()".
def presentar_quiz():
  
  global preguntas
  global pregunta

  contador_preguntas_totales_quiz = 0
  contador_preguntas_correctas_quiz = 0

  for pregunta in preguntas:
    
    tiempo = 1200
    inicio = time.time()

    número_de_pregunta, área_de_estudio, pregunta, respuesta1, respuesta2, respuesta3, respuesta4, correcta = pregunta # esto (pregunta) es una lista, hacemos asignación múltiple

    print(f"ID de la pregunta: {número_de_pregunta} \nEnfoque de la pregunta: {área_de_estudio} \nPregunta: {pregunta}")
    print(f"a){respuesta1:<15}b){respuesta2:<15}c){respuesta3:<15}d){respuesta4:<15} \n")

    contador_preguntas_totales_quiz = contador_preguntas_totales_quiz + 1

    respuesta_correcta = input("Ingresa la respuesta correcta: ")

    if respuesta_correcta == correcta:

      contador_preguntas_correctas_quiz = contador_preguntas_correctas_quiz + 1

    fin = time.time()

    tiempo_total = round((tiempo - (fin - inicio)) / 60, 2)
    print(f'\nEl tiempo restante son {tiempo_total} minutos\n')

  calificación = (contador_preguntas_correctas_quiz / contador_preguntas_totales_quiz) * 100
  reportar_calificaciones(calificación)


## Dentro de esta función se usa el valor de la función calificación dentro de la función "presentar_quiz()", asi poder mostrar el promedio obtenido, ussando la regla de 3.
# Se resta la calificación de 100 para obtener la calificación incorrecta.
# Usando print se muestra la calificación de 100 y el porcentaje correcto e incorrecto.
def reportar_calificaciones(calificación_quiz):

  menu()
  
  calificación = calificación_quiz

  porcentaje_incorrecto = 100 - calificación

  print(f'Tu calificación es {calificación}\nTu porcentaje de preguntas correctas es {calificación}%\nTu porcentaje de preguntas incorrectas es {porcentaje_incorrecto}%')


## Esta función se usa para finalizar el programa.
# Se cambia el valor del ciclo True a False, así la función "main()" se termina y solo se despliga el agradecimiento al usuario y un mensaje de despedida.
def salir():

  print("Gracias por usar el programa\n¡Hasta pronto!")

  continuar = False


def menu():

  print(f'\nMenú de opciones\n1. Alta de preguntas de prueba PISA (lectura, matemáticas, ciencias)\n2. Actualizar preguntas de la prueba PISA (lectura, matemáticas, ciencias)\n3. Estudiar preguntas de cierta área (lectura, matemáticas, ciencias)\n4. Presentar un quiz (lectura, matemáticas, ciencias)\n5. Reporte de calificaciones\n6. Salir\n')

  opción_menu = int(input("Elige el número de la opción deseada: "))
  print("\n")

  return opción_menu


def main():
  
  continuar = True

  while continuar:

    opción_menu = menu()

    if opción_menu < 1 or opción_menu > 6:

      print("Selecciona una de las opciones del menú")
      continue

    elif opción_menu == 1:
      registrar_pregunta()

    elif opción_menu == 2:
      actualizar_pregunta()

    elif opción_menu == 3:
      estudiar_área()

    elif opción_menu == 4:
      presentar_quiz()

    elif opción_menu == 5:
      reportar_calificaciones()

    elif opción_menu == 6:
      continuar = salir()

main()

## Reflexión
# Durante este proyecto, aprendí a manejar y a codificar con el lenguaje de Pyhton, utilizando todas las herramientas que se me otorgaron durante este curso.
# Poder entender como funcionan las computadoras y como es que se programan fue algo que aprendí al elaborar este proyecto, en donde yo pude realizar mi propio programa.
# 
```
